```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import os
import pickle
import re
from scipy.stats import wilcoxon
from sklearn.linear_model import LinearRegression
```

```python
sensibility = pd.read_csv('deterministic_sensitivity_analysis.csv')
sensibility
```

```python
sns.scatterplot(sensibility.groupby('tch_scale').agg({'objective_value': 'mean'}),x='tch_scale',y='objective_value')
```

```python
sns.scatterplot(sensibility.groupby('transp_scale').agg({'objective_value': 'mean'}),x='transp_scale',y='objective_value')
```
